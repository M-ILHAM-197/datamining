**Functions**

````python
def countCategory(data_column, label, column, category):
    group = data_column.groupby(category).count()
    amount_of_category = []
    for label in label:
        amount_of_category.append(group.loc[label, column])
    return amount_of_category
````

````python
def split(split_value, data_column, label, col, category):
    less_group = data_column[data_column[col] < split_value]
    greater_group = data_column[data_column[col] >= split_value]
    
    length_less_group = countCategory(less_group, label, col, category)
    length_greater_group = countCategory(greater_group, label, col, category)
    
    return (length_less_group, length_greater_group)
````

$$
Gain(E_{new}) = (E_{initial}) \ - (E_{new})
$$

````python
def count_gain(inisial_entropy, new_entropy):
  return inisial_entropy - new_entropy
````

````python
def count_entropy(data_target):
    all_prob = []
    for prob in data_target:
        if (prob/sum(data_target) != 0):
            all_prob.append(prob/sum(data_target) * log2(prob/sum(data_target)))
        else:
            all_prob.append(0)
    return -(sum(all_prob))
````

$$
Info_A(D) = \frac{|D_1|}{|D|} Entropy (D_1)\frac{|D_2|}{|D|}+ Entropy (D_2)
$$

````python
def info(d, data_target):
    temp = []
    for value in d:
        temp.append((sum(value) / data_target.shape[0]) * count_entropy(value))
    return sum(temp)
````


*  Hitung target entropy

*  Hitung target entropy dengan split value

*  Hitung entropy gain

````python
initial_data_target_entropy = count_entropy(countCategory(data_target, label, data_target.columns[0], data_target.columns[1]))
````

*  Nilai split yaitu 0.7

*  Data kurang dari < 0.7

*  Data kurang dari >= 0.7

````python
entropy_data_target_2 = info(split(0.7, data_target, label, data_target.columns[0], data_target.columns[1]), data_target)
````




*   Nilai split yaitu 1.4
*   Data kurang dari < 1.4
*   Data kurang dari >= 1.4




````python
entropy_data_target_1 = info(split(1.4, data_target, label, data_target.columns[0], data_target.columns[1]), data_target)
````
